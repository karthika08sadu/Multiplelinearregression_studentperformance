```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
df=pd.read_csv("Student_Performance.csv")
```

```python
df.head()
```

```python
df.drop("Extracurricular Activities",axis=1,inplace=True)
```

```python
df.head()
```

```python
df=df.rename({"Performance Index":"performance","Hours Studied":"hours","Previous Scores":"previous","Sleep Hours":"sleep","Sample Question Papers Practiced":"sample"},axis=1,inplace=True)
```

```python
df.columns = [
    'hours',
    'previous',
    'sleep',
    'sample',
    'performance'
]
```

```python
model=smf.ols("performance~hours+previous+sleep+sample",data=df).fit()
```
